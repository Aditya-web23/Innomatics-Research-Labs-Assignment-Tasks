```python
# Problem 1: Social Media – Post Engagement Analyzer
def analyze_post(likes):
    total_likes = 0
    for like in likes:
        total_likes += like

    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print("Total Likes:", total_likes)
    print("Post Status:", status)


likes = [200, 300, 250, 300]
analyze_post(likes)
```

```python
# Problem 2: Healthcare – Medicine Stock Alert System
def check_stock(stock):
    if stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"

    print("Medicine Stock:", stock)
    print("Status:", status)


check_stock(6)
```

```python
# Problem 3: Agriculture – Rainfall Adequacy Checker
def check_rainfall(rainfall, required_level):
    total = 0
    for r in rainfall:
        total += r

    average = total / len(rainfall)

    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Average Rainfall:", average)
    print("Rainfall Status:", status)


rainfall = [70, 75, 68, 75]
check_rainfall(rainfall, 70)
```

```python
# Problem 4: Social Media – Duplicate Account Detection
def check_duplicates(usernames):
    seen = set()
    duplicate_found = False

    for user in usernames:
        if user in seen:
            duplicate_found = True
            break
        seen.add(user)

    if duplicate_found:
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


users = ["john", "raj", "john"]
check_duplicates(users)
```

```python
# Problem 5: Healthcare – Appointment Eligibility Checker
def check_eligibility(age):
    if age >= 18:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Patient Age:", age)
    print("Eligibility Status:", status)


check_eligibility(21)
```

```python
# Problem 6: Agriculture – Premium Crop Price Filter
def filter_premium(prices):
    premium = []

    for price in prices:
        if price > 2000:
            premium.append(price)

    print("Premium Crops:", premium)


prices = [1500, 2500, 1800, 3200]
filter_premium(prices)
```

```python
# Problem 7: System Monitoring – Application Health Checker
def check_health(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print("Error Count:", errors)
    print("System Status:", status)


check_health(7)
```

```python
# Problem 8: Banking – Daily Transaction Limit Checker
def check_transaction(amount):
    limit = 50000

    if amount <= limit:
        status = "Approved"
    else:
        status = "Rejected"

    print("Transaction Amount:", amount)
    print("Transaction Status:", status)


check_transaction(60000)
```

```python
# Problem 9: E-Learning – Student Attendance Eligibility System
def check_attendance(attendance_list, total_days):
    present_days = 0
    for day in attendance_list:
        present_days += day

    percentage = (present_days / total_days) * 100

    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", status)


attendance = [1, 1, 1, 1, 0, 1, 1, 1]
check_attendance(attendance, 8)
```

```python
# Problem 10: Smart Electricity Bill Analyzer
def calculate_bill(units):
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    print("Total Bill: ₹", bill)
    print("Usage Level:", usage)


calculate_bill(250)
```
