```python
#Q2. Pass / Fail Analyzer
marks = [45, 78, 90, 33, 60]
pass_count=0
fail_count=0
for i in marks:
  if i>=50:
   pass_count+=1
  else:
   fail_count+=1
print("Toatal Pass Students count:",pass_count)
print(" Total Fail Students count:",fail_count)
```

```python
#Q3. Simple Data Cleaner
names = [" Alice ", "bob", " CHARLIE "]
cleaned_names = []

for name in names:
  cleaned_name = name.strip().lower() #used builtin methods for cleaning strip for space removing at end and start and lower for converting letters into lower character
  cleaned_names.append(cleaned_name)

print(cleaned_names)
```

```python
#4. Message Length Analyzer
messages = ["Hi", "Welcome to the platform", "OK"]
for message in messages:
    length = len(message)
    print("Message:", message)
    print("Length:", length)

    if length > 10:
        print("(Flag) This is long message")

    print() # this is for blank line to look our output clean format
```

```python
#Q5. Error Message Detector
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
count=0
for i in logs:
  if i=="ERROR":
   count+=1
print("Total Error Count:",count)
```
